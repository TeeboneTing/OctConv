```python
import tensorflow as tf
import numpy as np
# load dataset for POC use
import tensorflow.keras.datasets as keras_datasets
(x_train, y_train), (x_test, y_test) = keras_datasets.cifar10.load_data()
```

```python
# Normalize data
x_train = (x_train.astype(np.float32) - 127.5) / 127.5
x_test = (x_test.astype(np.float32) - 127.5) / 127.5

from tensorflow.keras.utils import to_categorical
y_train_one_hot = to_categorical(y_train)
```

```python
import sys
sys.path.append("OctConv-TFKeras")
from oct_conv2d import OctConv2D
from models import *
```

```python
normal_resnet = create_normal_wide_resnet()
normal_resnet.summary()
```

```python
from tensorflow.keras.optimizers import Adam, SGD
adam = Adam()
normal_resnet.compile(optimizer=adam, loss='binary_crossentropy', metrics=['accuracy'])
```

# Train normal_resnet


```python

normal_resnet.fit(x_train,y_train_one_hot,batch_size=64, epochs=10)
```

```python
normal_resnet.predict(x_test,verbose=1)
```
